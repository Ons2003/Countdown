```python
import time 
sec = int(input('Please input the length of the countdown in seconds:'))

def countdown(t) :
  while t != 0 :
    timeformat = divmod(t,60)
    print(timeformat[0],':',timeformat[1])
    time.sleep(1)
    t = t - 1
  print('Fire in the hole') 
  
#the end='\r' didn't work
countdown(sec)


```
